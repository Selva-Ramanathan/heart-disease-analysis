```python
#Heart disease - This was completed in the codeacademy editor and not here
# Heart disease part 1:

# import libraries
import codecademylib3
import pandas as pd
import numpy as np

# load data
heart = pd.read_csv('heart_disease.csv')
yes_hd = heart[heart.heart_disease == 'presence']
no_hd = heart[heart.heart_disease == 'absence']

#1.

chol_hd = yes_hd.chol

print('The average cholestrol level of patients with heart disease is: ', np.mean(chol_hd))

#the value is higher than 240 suggesting high colestrol levels for people with heart disease

from scipy.stats import ttest_1samp

chol_hd = yes_hd.chol

t_stat, pval = ttest_1samp(chol_hd, 240)

pval = pval/2 

print('pval of patients with heart disease: ', pval)

# the pval is less than 0.05 so we reject the null hypothesis

# testing for patients without heart disease
not_chol_hd = no_hd.chol

print('The average cholestrol level of patients without heart disease is: ', np.mean(not_chol_hd))

t_stat, pval = ttest_1samp(not_chol_hd, 240)

pval = pval/1

print('pval of patients without heart disease: ', pval)

#Since the pval is greater than the significance threshold of 0.05 we cannot say for certain if patients without heart disease have a higher average of cholestrol levels than 240mg/dl as we do not have enough evidence. 

num_patients = len(heart)

print(num_patients)

num_highfbs_patients = np.sum(heart.fbs)

print(num_highfbs_patients)

num_expected_diabetes = 0.08 * num_patients
print(num_expected_diabetes)

# the value from the general population (8%) is 24.24 which is less than what I got from the data set heart, 45. The data set targets people who have or are being evaluated for heart disease whereas the general population is random people who may or may not have heart disease. People with heart disease will more likely have high blood sugar, high cholestrol and diabetes. 
 

from scipy.stats import binom_test
pval= binom_test(num_highfbs_patients, num_patients, 0.08, alternative='greater')

print(pval)

# As this value is less than the significance, we'll reject the null hypothesis and the sample is from a population that has an FBS > 120 which is higher than 8%. 


```
